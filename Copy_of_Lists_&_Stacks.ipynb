# Java Collections Framework

The Java Collections Framework supports two types of containers (data structures):

* Collections - used for storing a collection of elements
> * **Lists**
> * **Stacks**
> * Sets
> * Queues
> * Priority Queues

* Maps - used for storing key/value pairs

These notes focus on Lists and Stacks. The [Collection interface](https://docs.oracle.com/javase/8/docs/api/java/util/Collection.html) defines the common operations for lists, vectors, stacks, queues, priority queues and sets.

# Iterable & Iterator

Each collection is [Iterable](https://docs.oracle.com/javase/8/docs/api/java/lang/Iterable.html). You can obtain its [Iterator](https://docs.oracle.com/javase/8/docs/api/java/util/Iterator.html) object to traverse all the elements in the collection. Iterator is a classic design pattern for walking through a data structure without having to expose the details of how data is stored in the data structure.

The Collection interface extends the Iterable interface. The Iterable interface defines the iterator method, which returns an iterator. Thus, the iterator method must return an instance of an Iterator. Consequently, a custom class that implements the Iterator interface must be defined. This custom class must implement the hasNext and next methods such that the data in a data structure is traverse progressively from beginning to end.

A class that successfully implements the Iterable interface, as all Collections do, can be the target of the forEach method. The forEach method is a default method added in Java 8.

# Lists

The [List interface](https://docs.oracle.com/javase/8/docs/api/java/util/List.html) extends the Collection interface and defines a collection for storing elements in a sequential order. To create a list, use one of its two concrete classes: [ArrayList](https://docs.oracle.com/javase/8/docs/api/java/util/ArrayList.html) or [LinkedList](https://docs.oracle.com/javase/7/docs/api/java/util/LinkedList.html). 

The List interface allows a collection of elements to be stored in sequential order and duplicate elements are allowed. The List interface adds position-oriented operations. Thus, elements in a list are often identified by their index value. The following methods are defined in the List interface:

* add( index, element )
* addAll( index, collection )
* get( index )
* indexOf( element )
* lastIndexOf( element )
* listIterator( )
* listIterator( startIndex )
* remove( index )
* subList( fromIndex, toIndex )
* set( index, element )

## [ArrayList](https://docs.oracle.com/javase/8/docs/api/java/util/ArrayList.html)

The ArrayList class stores elements in an array. The array is dynamically created, if the capacity of the array is exceeded, a larger new array is created and all the elements from the current array are copied to the new array.

ArrayList is a resizable-array implementation of the List interface. It also provides methods for manipulating the size of the array used internally to store the list. The trimToSize method can be called to trim the size of the array to match the current size of the list. 

## [LinkedList](https://docs.oracle.com/javase/7/docs/api/java/util/LinkedList.html)

LinkedList stores elements in a linked list. A linked list provides the methods for retrieving, inserting, and removing elements from both ends of the list (first & last). The following methods are implemented by the LinkedList class:

* addFirst( element )
* addLast( element )
* getFirst( )
* getLast( )
* removeFirst( )
* removeLast( )

## ArrayList vs LinkedList

Which of the two classes you use depends on your specific needs. The critical difference between them pertains to the internal implementation, which affects their performance. LinkedList is efficient for inserting and removing elements at the beginning of the list, and ArrayList is more efficient for all other operations. The get(i) method is available for a linked list, but it is a time-consuming operation. Do not use it to traverse all the elements in a list. Instead, use a for-each loop. 

# The Comparator Interface

The [Comparator interface](https://docs.oracle.com/javase/8/docs/api/java/util/Comparator.html) can be used to compare the objects of a class that doesn't implement the Comparable interface. Comparator can also be used to define a new criteria for comparing objects.

You can define a comparator to compare the elements of different classes. To do so, define a class that implements the java.util.Comparator<T\> interface and overrides the compare method.

> public int compare( T element1, T element2 )

The compare method returns a negative value if element1 is less than element2, a positive value if element1 is greater than element2, and zero if they are equal.

Custom classes can be defined to compare elements based on different criteria. Each class would implement the compare method according to the criteria to be used.

The example above provides an example of how to implement the Comparator interface.

```java
public class SortingStrings
{
   public static void main(String[] args)
   {
      String[] cities = {"Atlanta", "Savannah", "New York", "Dallas"};
      java.util.Arrays.sort( cities, new LengthComparator() ); //sorts in ascending string length order
      
      for( String s : cities )
         System.out.print( s.length() + ": " + s + "  " );
         
      System.out.println();
      
      java.util.Arrays.sort( cities, new AlphaComparator() );  //sorts in alphabetical order
      
      for( String s : cities )
         System.out.print( s + "  " );
   }
   
   static class LengthComparator implements java.util.Comparator<String>
   {
      @Override
      public int compare( String s1, String s2 )
      {
         return s1.length() - s2.length();
      }
   }
   
   static class AlphaComparator implements java.util.Comparator<String>
   {
      @Override
      public int compare( String s1, String s2 )
      {
         return s1.compareTo(s2);
      }
   }
}
```

# Vector & Stack Classes

[Vector](https://docs.oracle.com/javase/8/docs/api/java/util/Vector.html) is a subclass of [AbstractList](https://docs.oracle.com/javase/8/docs/api/java/util/AbstractList.html) and [Stack](https://docs.oracle.com/javase/8/docs/api/java/util/Stack.html) is a subclass of Vector in the Java API. Vector is the same as ArrayList, except that it contains synchronized methods for accessing and modifying the vector. Synchronized methods can prevent data corruption when a vector is accessed or modified by two or more threads simultaneously. Synchronization will likely be the topic of future courses. In all, if your application does not require synchronization, using ArrayList is more efficient that using Vector. 

The Stack class is a last-in, first-out data structure. It's push method permits the addition of an element to the top of the stack, while its pop method removes an element from the top of the stack. Although the Stack class has a get method, like the LinkedList, it is not a good idea to access elements from the stack with it. 

# Read File
